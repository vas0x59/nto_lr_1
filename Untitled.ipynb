```python
def d(l1: LineSP, l2: LineSP):
    return np.linalg.norm(l1.p_tau(0.5) - l2.p_tau(0.5))
```

```python
lines = [
    LineSP(v=np.array([1, 1]), p0=np.array([0, 0]), t_max=1),
    LineSP(v=np.array([1, 2]), p0=np.array([0, 0]), t_max=1),
    LineSP(v=np.array([1, 2]), p0=np.array([1, 1]), t_max=1),
    LineSP(v=np.array([-1, 1]), p0=np.array([1, 1]), t_max=1),
    LineSP(v=np.array([1, 1]), p0=np.array([10, 10]), t_max=1),
    LineSP(v=np.array([1, 2]), p0=np.array([10, 10]), t_max=1),
    LineSP(v=np.array([1, 2]), p0=np.array([11, 11]), t_max=1),
    LineSP(v=np.array([0.5, 2]), p0=np.array([11, 11]), t_max=1),
    LineSP(v=np.array([-1, 2]), p0=np.array([101, 101]), t_max=1)
]
```

```python
X = np.array([
    [*l.p0, *l.v, l.t_max]
    for l in lines
])
```

```python
rect_overlap_metric(lines[3], lines[4])
```

```python
def metric(x, y):
    return simple_distance(LineSP(x[:2], x[2:4], x[4]), LineSP(y[:2], y[2:4], y[4]))
```

```python
clustering = DBSCAN(eps=0.7, min_samples=1, metric=metric).fit(X)
```

```python
n = len(set(clustering.labels_))
```
